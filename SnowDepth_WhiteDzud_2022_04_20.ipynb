```python
# installing dependencies
#
import pandas as pd
import os
import seaborn as sns
import numpy as np
import datetime
```

```python
# read in CSV file for Umnugovi (all years)
#
snow_depth = pd.read_csv ("D:/Users/aster/Desktop/SNOW_DEPTH_2022_04_20/Snow_Depth_all_data_clean.csv")
df = pd.DataFrame(snow_depth)
df
```

```python
df.dtypes
```

```python
#Organize in alphabetical order
#
df = df.reindex(sorted(df.columns), axis=1)
df
```

```python
#Add DateTime column with Temperature at 3 hr intervals
#
df['DateTime'] = pd.date_range(start='2000-01-01', periods=len(df), freq='3H')
df
```

```python
df.dtypes
```

```python
#Set DateTime as index
#
df= df.set_index('DateTime')
df
```

```python
#Save as CSV file
#
df.to_csv('SnowDepth_WhiteDzud_3hr.csv', index=True)
print('new csv file from list was created... ')
```

#### 28 day moving average for selected time interval

```python
run_mean = df['AirTemp'].rolling(28, center=True).mean()
run_mean
```

```python
# removing all the NULL values using 
# dropna() method
run_mean.dropna(inplace=True)

# using .to_frame() to convert pandas series
# into dataframe.
run_mean = run_mean.to_frame()
run_mean
```

```python
run_mean.plot(title='28-day running mean', ylabel='Temperature [℃]', color= 'sandybrown',
								figsize=(16, 8))
```

```python
#Combine 2 dataframes
#
frames = [df, run_mean]
combined = pd.concat(frames, axis=1)

#save as csv
#
combined.to_csv('combined_July_2009-2011')

#show
#
combined
```

```python
#Plot
#
combined.plot(title='Daily Mean & Moving Average 28 days 09-2009 to 04-2010', ylabel='Temperature [℃]',
								figsize=(16, 8))
```
